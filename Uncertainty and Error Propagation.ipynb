```python
#Analytical methods and Monte Carlo simulation were applied in this project.
import numpy as np
import matplotlib.pyplot as plt
```

```python
r_basic=12.0
r_alpha=0.5
N=50000
```

```python
def area(r):
    return np.pi*r**2
```

```python
#Analitik Error Propagation
da_dr=2*np.pi*r_basic
alpha_A_analytic=da_dr*r_alpha
A_basic=area(r_basic)
```

```python
#Monte Carlo Sim√ºlayonu
r_examples=np.random.normal(r_basic, r_alpha, N)
A_examples=area(r_examples)
```

```python
A_mc_basic=np.mean(A_examples)
A_mc_std=np.std(A_examples)
```

```python
difference=abs(A_mc_std-alpha_A_analytic)
```

```python
plt.hist(A_examples, bins=50, density =True)
plt.show()
```
